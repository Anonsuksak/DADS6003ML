```python
data = pd.read_csv("https://raw.githubusercontent.com/Anonsuksak/DADS6003ML/main/Housing.csv")
data
```

## 1.1 Look if there are any duplicate rows in the dataset

## 1.2 Check for null values

```python
data.isna().sum()
```

- No missing values

## 1.3 เราทราบว่า มี string 7 ตัวแปร โดยที่มีคำตอบเป็น yes, no 6 ตัวแปร อีกตัวแปร furnishingstatus ไม่ใช่ yes, no จึงทำ groupby เพื่อดูข้อมูลว่ามีอะไรบ้าง สำหรับแปลงให้เป็น dummy ขั้นต่อไป

```python
data.info()
```

```python
data.groupby("furnishingstatus")["furnishingstatus"].count()
```

## 1.4 แปลง string ให้เป็น dummy (int) เพื่อวิเคราะห์ขั้นต่อไป

```python
data['mainroad'] = data['mainroad'].replace({'yes':1,'no':0}).astype(int)
data['guestroom'] = data['guestroom'].replace({'yes':1,'no':0}).astype(int)
data['basement'] = data['basement'].replace({'yes':1,'no':0}).astype(int)
data['hotwaterheating'] = data['hotwaterheating'].replace({'yes':1,'no':0}).astype(int)
data['airconditioning'] = data['airconditioning'].replace({'yes':1,'no':0}).astype(int)
data['prefarea'] = data['prefarea'].replace({'yes':1,'no':0}).astype(int)
data['furnishingstatus'] = data['furnishingstatus'].replace({'furnished':2,'semi-furnished':1, 'unfurnished':0 }).astype(int)
```

```python
data
```

## Step 2 ML Model

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x  = data.loc[:,"area":]
y = data["price"]
```

- อีกโค้ดที่สามารถกำหนดตัวแปร x, y
- X = data.drop(['price'],axis=1)
- y = data['price']

```python
scalar = StandardScaler() ##สร้างอินสแตนซ์ของ StandardScaler
x_scale = scalar.fit_transform(x) ##ปรับค่าของคุณลักษณะในชุด Train
```

## 2.1 Train and test split data เป็นสัดส่วน 80:20 โดย 80 สำหรับ Train process และ 20 สำหรับ Test process

```python
x_train, x_test, y_train, y_test = train_test_split(x_scale,y, test_size= 0.2 , random_state=(99))
```

## 2.2 สร้าง ML regression model และทดสอบ MSE กับ testset (แสดงค่า mse)

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
columnlist = list(data.columns)

# ผสานข้อมูลจาก feature_names และ coefficients ด้วยฟังก์ชัน zip()
coefficients_list = list(zip(columnlist, reg.coef_))

# แสดง intercept
print('Intercept:', reg.intercept_)

# แสดง coefficients โดยใช้ลูป
for feature, coefficient in coefficients_list:
    print(feature + ':', coefficient)
```

```python
y_pred_reg = reg.predict(x_test)
```

```python
import numpy as np
from sklearn import metrics # Evaluation for Linear Regression
from sklearn.metrics import r2_score # Linear Regression Score


print('MSE:', metrics.mean_squared_error(y_test, y_pred_reg))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred_reg)))
print('R sqaure:', r2_score(y_test, y_pred_reg))
```
